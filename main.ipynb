```python
import time
t = time.time()
import pandas as pd
print("Import Pandas:", round(time.time() - t, 3), "sec")

df = pd.read_csv('amazon.csv')
df = df.dropna()
print(df.head())
```

```python
#Information generale sur le dataset

df_shape = df.shape
print("Nombre de lignes du dataset:", df_shape[0])
print("Nombre de colonnes:", df_shape[1])

df.dtypes
```

```python
#Changement de type dans les colonnes numeriques

df["discounted_price"] = (df["discounted_price"].str.replace("‚Çπ", "", regex=False).str.replace("," , "", regex=False).astype(float))

df["actual_price"] = (
    df["actual_price"]
    .str.replace("‚Çπ", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
#Discount

df["discount_percentage"] = (
    df["discount_percentage"].str.replace("%", "", regex=False).astype(float)
)
```

```python
#Ratings
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df["rating_count"] = (
    df["rating_count"]
    .str.replace(",", "", regex=False)
    .astype(int)
)
```

```python
print(df[["actual_price","discounted_price","discount_percentage", "rating", "rating_count"]].head())
```

```python
#Verification des doublons 

print("Lignes repetees",df.duplicated().sum())
print("Reviews repetees", df['review_id'].duplicated().sum())
print("Produits repetees",df['product_id'].duplicated().sum())
```
